```python
#Importing dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading a csv file
data_file = "Resources/country_vaccinations_by_manufacturer.csv"
data_file_df = pd.read_csv(data_file)
data_file_df
```

```python
# type of vaccines
type_of_vaccine = data_file_df.groupby("vaccine")["total_vaccinations"].count()
vaccines_by_country = data_file_df.groupby("location")["total_vaccinations"].sum()
vaccines_by_country_df = pd.DataFrame({"Total no. of Vaccinations": vaccines_by_country})
type_of_vaccine
vaccines_by_country_df
```
